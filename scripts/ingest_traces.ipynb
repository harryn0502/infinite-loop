```python
import json
from ingestion import ingest_trace

json_path = "../raw_data/langsmith_traces.json"  # adjust if needed

with open(json_path, "r") as f:
    data = json.load(f)

# Handle both list and single-object cases safely
if isinstance(data, dict):
    traces = [data]
elif isinstance(data, list):
    traces = data
else:
    raise TypeError(f"Unexpected JSON root type: {type(data)}")

for trace in traces:
    ingest_trace(trace)

print(f"Ingested {len(traces)} traces into SQLite.")
```
